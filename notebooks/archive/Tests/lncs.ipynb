```python
# Downloads the html-Code of the given URL

import urllib.request, urllib.error, urllib.parse

#url = 'https://link.springer.com/book/10.1007/978-3-031-07469-1'

url = input('Enter URL to download: ')

response = urllib.request.urlopen(url)
webContent = response.read().decode('UTF-8')

#print(webContent[0:300]) #print first lines of html code for testing reasons
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webContent, 'html.parser')

print(soup.prettify()) #prints a more prettier version of the html code

soup.title.string
```

```python
import re #for using regular expressions by finding the paper titles

titles = soup.find_all(href = re.compile('/chapter/'))

for title in titles:
    print(title.string)
```

```python
authors = soup.find_all(class_ = 'c-author-list__item')
for author in authors:
    print(author.string)
```

```python
with open('ICAA_2022.csv', 'w') as f:
    f.write('Venue;Title;Authors;Abstract\n')
    for i in range(len(titles)):
        f.write('ICAA 2022;')
        f.write(titles[i].string)
        f.write(';')
        f.write(authors[i].string)
        f.write('\n')
```
